```python
import json
import re

def unescape_keys(obj):
    if isinstance(obj, dict):
        new_dict = {}
        for key, value in obj.items():
            # Only unescape \uXXXX patterns, leave other escapes alone
            unescaped_key = re.sub(
                r'\\u([0-9a-fA-F]{4})',
                lambda m: chr(int(m.group(1), 16)),
                key
            )
            
            # Recursively process the value
            new_dict[unescaped_key] = unescape_keys(value)
        return new_dict
    elif isinstance(obj, list):
        return [unescape_keys(item) for item in obj]
    else:
        # Keep values unchanged
        return obj

with open('Gemma3TokenizerFast.json', 'r', encoding='utf-8-sig') as f:
    data = json.load(f)

# Unescape only keys
unescaped_data = unescape_keys(data)

with open('Gemma3TokenizerFastUnescaped.json', 'w', encoding='utf-8', errors='surrogatepass') as f:
    json.dump(unescaped_data, f, ensure_ascii=False, indent=2)

print("Successfully unescaped keys in tokenizer file!")


# note 245237 (MASK IT SOMEHOW) because it is equal to space.
```
